```python
# STEP 2: LOAD AND INSPECT DATA

# ðŸ“¦ Import required libraries
import pandas as pd

# ðŸ“ Load dataset
df = pd.read_csv('../data/drugs_side_effects_drugs_com.csv')

# ðŸ§¾ Show first 5 rows
print("ðŸ”¹ First 5 rows:")
display(df.head())

# ðŸ”¢ Dataset shape
print(f"\nðŸ”¸ Dataset has {df.shape[0]} rows and {df.shape[1]} columns.")

# ðŸ·ï¸ Column names
print("\nðŸ”¹ Columns in the dataset:")
for col in df.columns:
    print(f" - {col}")

# ðŸ§  Dataset info
print("\nðŸ”¸ Dataset Info:")
display(df.info())

# ðŸ§¼ Missing values summary
print("\nðŸ”¸ Missing values per column:")
missing = df.isnull().sum()
display(missing[missing > 0])  # Only show columns that have missing values

# âœ… Optional: show % of missing data per column
print("\nðŸ“Š % of Missing Data per Column:")
missing_percent = (df.isnull().sum() / len(df)) * 100
display(missing_percent[missing_percent > 0].round(2))
```

```python
# STEP 3: CLEANING & PREPROCESSING

import numpy as np

# ðŸ’§ Fill missing values for text columns with "Unknown"
df['side_effects'] = df['side_effects'].fillna('Unknown')
df['related_drugs'] = df['related_drugs'].fillna('Unknown')
df['generic_name'] = df['generic_name'].fillna('Unknown')
df['drug_classes'] = df['drug_classes'].fillna('Unknown')
df['rx_otc'] = df['rx_otc'].fillna('Unknown')
df['pregnancy_category'] = df['pregnancy_category'].fillna('Unknown')

# ðŸ”¢ Fill missing values for rating and review count with 0
df['rating'] = df['rating'].fillna(0)
df['no_of_reviews'] = df['no_of_reviews'].fillna(0)

# ðŸ’Š Clean the 'activity' column (remove % and convert to float)
df['activity'] = df['activity'].astype(str).str.replace('%', '').str.strip()
df['activity'] = pd.to_numeric(df['activity'], errors='coerce') / 100

# ðŸ· Convert alcohol column: X = interacts (1), NaN = no interaction (0)
df['alcohol'] = df['alcohol'].replace({'X': 1, np.nan: 0})

# ðŸ” Check for any remaining missing values
print("âœ… Remaining missing values after cleaning:")
display(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# 1. Rating Distribution
plt.figure()
sns.histplot(df['rating'], bins=10, kde=True, color='teal')
plt.title('Distribution of Drug Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Drugs')
plt.show()

# 2. Top 10 Most Reviewed Drugs
top_reviewed = df[['drug_name', 'no_of_reviews']].groupby('drug_name').sum().sort_values(by='no_of_reviews', ascending=False).head(10)

plt.figure()
sns.barplot(x=top_reviewed['no_of_reviews'], y=top_reviewed.index, palette='viridis')
plt.title('Top 10 Most Reviewed Drugs')
plt.xlabel('Number of Reviews')
plt.ylabel('Drug Name')
plt.show()
```

```python
# 3. Top 10 Most Common Medical Conditions
top_conditions = df['medical_condition'].value_counts().head(10)

plt.figure()
sns.barplot(x=top_conditions.values, y=top_conditions.index, palette='magma')
plt.title('Top 10 Most Common Medical Conditions')
plt.xlabel('Count')
plt.ylabel('Medical Condition')
plt.show()

# 4. Top 10 Most Frequent Side Effects
side_effect_series = df['side_effects'].str.split(';').explode().str.strip()
top_side_effects = side_effect_series.value_counts().head(10)

plt.figure()
sns.barplot(x=top_side_effects.values, y=top_side_effects.index, palette='coolwarm')
plt.title('Top 10 Most Common Side Effects')
plt.xlabel('Count')
plt.ylabel('Side Effect')
plt.show()
```

```python
# STEP 5: Create Clean Dashboard-Ready DataFrame

dashboard_df = df[[
    'drug_name',
    'medical_condition',
    'rating',
    'no_of_reviews',
    'side_effects',
    'activity',
    'rx_otc',
    'pregnancy_category',
    'csa',
    'alcohol'
]].copy()

# Save cleaned version to file
dashboard_df.to_csv('../data/cleaned_drugs_dashboard.csv', index=False)

print("âœ… Cleaned data saved to: data/cleaned_drugs_dashboard.csv")
display(dashboard_df.head())
```
